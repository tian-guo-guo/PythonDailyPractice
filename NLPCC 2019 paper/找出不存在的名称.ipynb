```python
import requests
from lxml import etree

result = requests.get('http://tcci.ccf.org.cn/conference/2019/pro_session.php').content
html = etree.HTML(result)
names = html.xpath("//div[@class='programtitle']/a/span")
print(len(names)) # 解析出来176个标题
```

```python
len(set(names))
```
