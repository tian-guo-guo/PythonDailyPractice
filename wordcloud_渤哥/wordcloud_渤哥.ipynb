```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = open('boge1.txt').read()
font = r'User/suntian//Fonts/SimHei.ttf' 
wc = WordCloud(collocations=False,font_path=font,width=1400,height=1400,margin=2).generate(text.lower())
plt.imshow(wc)
plt.axis("off")
plt.show()
wc.to_file('1.png')
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

text = open('boge2.txt').read()
font = r'User/suntian//Fonts/SimHei.ttf' 
boge_coloring = np.array(Image.open("boge.png"))

# wc = WordCloud().generate(text.lower())
wc = WordCloud(background_color="white",collocations=False,font_path=font,width=1400,height=1400,max_font_size=40,margin=2,mask=boge_coloring)
# wc = WordCloud(background_color="white",font_path=font, max_words=2000, mask=alice_coloring, max_font_size=40, random_state=42)
wc.generate(text)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure()

plt.imshow(boge_coloring, cmap=plt.cm.gray, interpolation="bilinear")
plt.axis("off")
plt.show()
wc.to_file('pic.png')
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

text = open('boge22.txt').read()
font = r'User/suntian//Fonts/SimHei.ttf' 
boge_coloring = np.array(Image.open("boge22.png"))

wc = WordCloud(background_color="white",collocations=True,font_path=font,width=1400,height=1400,margin=2,mask=boge_coloring)
wc.generate(text)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure()

plt.imshow(boge_coloring, interpolation="bilinear")
plt.axis("off")
plt.show()
wc.to_file('boge2.png')
```
